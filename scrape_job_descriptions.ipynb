```python
import csv
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from webdriver_manager.chrome import ChromeDriverManager

# Initialize the Chrome Driver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))


# Function to scrape a single job page
def scrape_job_details(job_url):
    driver.get(job_url)
    wait = WebDriverWait(driver, 10)

    title = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="content"]/main/section[1]/div/div[1]/div[1]/h1')))
    requirements = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="content"]/main/section[2]/div/div/div[1]')))
    try:
        salary = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="content"]/main/section[2]/div/div/div[1]/div[2]/p[4]')))
    except NoSuchElementException:
        salary = "Not Listed"

    # Scrape the data Title -> Requiremenets -> Salary
#    scraped_data = [title.text, requirements.text, salary if isinstance(salary, str) else salary.text]
    scraped_data = [title.text, requirements.text]
    # Go back to the list page
    driver.back()
    time.sleep(3)  # Wait for 3 seconds for the page to load

    return scraped_data


# Function to get all job URLs from a page
def get_job_urls_from_page():
    wait = WebDriverWait(driver, 10)
    job_links = []
    MAX_JOBS_PER_PAGE = 10  # Adjust this number based on the website's layout

    for i in range(1, MAX_JOBS_PER_PAGE + 1):
        try:
            xpath = f'//*[@id="content"]/section/div/div/div/div[1]/div/div[{i}]/div/div[1]/div[2]/div/h4/a'
            job_link = wait.until(EC.presence_of_element_located((By.XPATH, xpath)))
            job_links.append(job_link.get_attribute('href'))
        except NoSuchElementException:
            break

    return job_links

# Start with the first page of job listings
BASE_URL = "https://datasciencejobs.com/jobs/"
MAX_PAGES = 76

# Prepare CSV file
with open('job_descriptions.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Title', 'Descriptions'])
    for page in range(1, MAX_PAGES + 1):
        page_url = f"{BASE_URL}page/{page}/" if page > 1 else BASE_URL
        driver.get(page_url)
        time.sleep(3)  # Wait for the page to load

        urls = get_job_urls_from_page()
        for url in urls:
            try:
                job_detail = scrape_job_details(url)
                writer.writerow(job_detail)
            except Exception as e:
                print(f"Error scraping {url}: {e}")

'''# Prepare CSV file
with open('job_details.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Title', 'Requirements', 'Salary'])

    for page in range(1, MAX_PAGES + 1):
        page_url = f"{BASE_URL}page/{page}/" if page > 1 else BASE_URL
        driver.get(page_url)
        time.sleep(3)  # Wait for the page to load

        urls = get_job_urls_from_page()
        for url in urls:
            try:
                job_detail = scrape_job_details(url)
                writer.writerow(job_detail)
            except Exception as e:
                print(f"Error scraping {url}: {e}")
'''
# Close the driver
driver.quit()

print("Scraping completed and data saved to job_details.csv")
```
